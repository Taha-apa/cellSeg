```python
import matplotlib.pyplot as plt
test = '/content/PBC_dataset_normal_DIB/neutrophil/BNE_104344.jpg'
result = model(test)
result[0].show()
```

```python
import torch
cls = 'WBC'
id = torch.where(result[0].boxes.cls == 2)[0]
print(id)
wbc_bbx = result[0].boxes[id].xyxy
print(wbc_bbx)
wbc_bbx = torch.Tensor.int(torch.floor(wbc_bbx))
print(wbc_bbx)
```

```python
def crop(box, i):
    x0 = box[0]
    y0 =  box[1]
    x1 = box[2]
    y1 = box[3]
    return i[y0:y1,x0:x1 , :]
```

```python
classes= ['basophil','eosinophil','lymphocyte','monocyte','neutrophil']
base_dir = "/content/PBC_dataset_normal_DIB"
num_class = len(classes)
image_files = [
    [os.path.join(base_dir, classes[i] +"/" + x) for x in os.listdir(os.path.join(base_dir, classes[i]))]
    for i in range(num_class)
]
num_each = [len(image_files[i]) for i in range(num_class)]
image_files_list = []
image_class = []
for i in range(num_class):
    image_files_list.extend(image_files[i])
    image_class.extend([i] * num_each[i])
num_total = len(image_class)
```

```python
!rm -rf /content/PBC_prepd
```

```python
!mkdir PBC_prepd
```

```python
import os
for cls in classes :
  os.mkdir(os.path.join('/content/PBC_prepd',cls))
```

```python
from PIL import Image
for index in range(num_total):
  cls = image_class[index]
  img_file = image_files_list[index]
  img = os.path.join(classes[cls],img_file)
  result = model(img)
  ids = torch.where(result[0].boxes.cls == 2)[0]
  print(ids)
  if len(ids) != 0 :
    for id in ids:
      rslt = result[0]
      bx = rslt.boxes[id]
      wbc_bbxs = bx.xyxy
      for wbc_bbx in wbc_bbxs:
        wbc_bbx = torch.Tensor.int(wbc_bbx)
        img_nd = np.array(PIL.Image.open(img))
        crp_img = crop(wbc_bbx,img_nd)
        crp_img = Image.fromarray(crp_img)
        img_path = '/content/PBC_prepd/' + classes[cls]+'/'+ str(index)+f'_{id}' + '.jpg'
        crp_img.save(img_path)
```
